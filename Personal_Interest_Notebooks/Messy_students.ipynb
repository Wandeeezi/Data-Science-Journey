```python
import pandas as pd
import warnings


warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv(r"C:\Users\MC\Documents\Projects\Kaggle\Datasets\messy_students.csv")
```

```python
df.info()
```

```python
df["Age"].replace("twenty-two","22",inplace=True)
```

```python
my_dict = {
    "Y" : "Yes",
    "yes" : "Yes",
    "N" :"No",
    "no" : "No",
    "YES" : "Yes"
}

df["Enrolled"].replace(my_dict,inplace=True)
```

```python
my_date_dict = {
    "05/02/2025"  :  "2025-05-02",
    "2025/03/10"   : "2025-03-10",
    "2025/06/20"   : "2025-06-20",
    "20-07-2025"  :  "2025-07-20",
    "2025/08/15"  :  "2025-08-15",
    "2025/09/01"  :  "2025-09-01"
}

df["SignupDate"].replace(my_date_dict,inplace=True)
```

```python
df["Age"] = pd.to_numeric(df["Age"],errors="coerce")
```

```python
df["Age"].fillna(df["Age"].mean(),inplace=True)
```

```python
df.head(10)
```

```python
df.sort_values(by="Marks",na_position="last",inplace=True)
```

```python
df=df.drop_duplicates(subset=["Name","ID"],keep="first")
```

```python
df.head(10)
```

```python
df["Marks"]= df["Marks"].clip(upper=100)
```

```python
df.head(10)
```

```python
df["City"] = df["City"].str.capitalize()
```

```python
df.head(10)
```

```python
df["Enrolled"].fillna("Unknown",inplace=True)
```

```python
grade_dict = {
    "A": (80, 100),
    "B": (70, 79), 
    "C": (60, 69), 
    "D": (50, 59),
    "E": (40, 49),
    "F": (0, 39) 
}

def assign_grade(marks):
    for grade, (low, high) in grade_dict.items():
        if low <= marks <= high:
            return grade
    return "Invalid" 
```

```python

df["Grade"] = df["Marks"].apply(assign_grade)
```
